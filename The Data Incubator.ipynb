```python
move = [0]
def tour_hanoi(disc):
    if (disc != 0):
        print(f'disc: {disc} position: {pos} move: {move} ll \n')
        if(pos[disc] == 0):
            tour_hanoi(disc - 1)
        else:
            print(f'disc: {disc} position: {pos} move: {move} 1 {T0} \n')
            if(len(T0)>=1 & T0[-1]>disc):
                print(f'disc: {disc} position: {pos} move: {move} l \n')
                if(len(T1)>=1 & T1[-1]==disc):
                    pos[disc]=0
                    T0.append(disc)
                    T1.remove(disc)
                    move[0] += 1
                    print(f'disc: {disc} position: {pos} move: {move} T0 \n')
                    tour_hanoi(disc-1)
                elif(len(T2)>=1 & T2[-1]==disc):
                    pos[disc]=0
                    T0.append(disc)
                    T2.remove(disc)
                    move[0] += 1
                    print(f'disc: {disc} position: {pos} move: {move} T0 \n')
                    tour_hanoi(disc-1)
                elif(len(T3)>=1 & T3[-1]==disc):
                    pos[disc]=0
                    T0.append(disc)
                    T3.remove(disc)
                    move[0] += 1
                    print(f'disc: {disc} position: {pos} move: {move} T0 \n')
                    tour_hanoi(disc-1)
                else:
                    tour_hanoi(disc-1)
            # Tower 1
            elif(len(T1)>=0 & T1[-1]>disc):
                print(f'disc: {disc} position: {pos} move: {move} l \n')
                if(len(T0)>=1 & T0[-1]==disc):
                    pos[disc]=0
                    T1.append(disc)
                    T0.remove(disc)
                    move[0] += 1
                    print(f'disc: {disc} position: {pos} move: {move} \n')
                    tour_hanoi(disc-1)
                elif(len(T2)>=1 & T2[-1]==disc):
                    pos[disc]=0
                    T1.append(disc)
                    T2.remove(disc)
                    move[0] += 1
                    print(f'disc: {disc} position: {pos} move: {move} \n')
                    tour_hanoi(disc-1)
                elif(len(T3)>=1 & T3[-1]==disc):
                    pos[disc]=0
                    T1.append(disc)
                    T3.remove(disc)
                    move[0] += 1
                    print(f'disc: {disc} position: {pos} move: {move} \n')
                    tour_hanoi(disc-1)
                else:
                    tour_hanoi(disc-1)
            # Tower 2
            elif(len(T2)>=0 & T2[-1]>disc):
                print(f'disc: {disc} position: {pos} move: {move} l \n')
                if(len(T0)>=1 & T0[-1]==disc):
                    pos[disc]=0
                    T2.append(disc)
                    T0.remove(disc)
                    move[0] += 1
                    print(f'disc: {disc} position: {pos} move: {move} \n')
                    tour_hanoi(disc-1)
                elif(len(T1)>=1 & T1[-1]==disc):
                    pos[disc]=0
                    T2.append(disc)
                    T1.remove(disc)
                    move[0] += 1
                    print(f'disc: {disc} position: {pos} move: {move} \n')
                    tour_hanoi(disc-1)
                elif(len(T3)>=1 & T3[-1]==disc):
                    pos[disc]=0
                    T2.append(disc)
                    T3.remove(disc)
                    move[0] += 1
                    print(f'disc: {disc} position: {pos} move: {move} \n')
                    tour_hanoi(disc-1)
                else:
                    tour_hanoi(disc-1)
            # Tower 3
            elif(len(T3)>0 & T3[-1]>=disc):
                print(f'disc: {disc} position: {pos} move: {move} l \n')
                if(len(T0)>=1 & T0[-1]==disc):
                    pos[disc]=0
                    T3.append(disc)
                    T0.remove(disc)
                    move[0] += 1
                    print(f'disc: {disc} position: {pos} move: {move} \n')
                    tour_hanoi(disc-1)
                elif(len(T1)>=1 & T1[-1]==disc):
                    pos[disc]=0
                    T3.append(disc)
                    T1.remove(disc)
                    move[0] += 1
                    print(f'disc: {disc} position: {pos} move: {move} \n')
                    tour_hanoi(disc-1)
                elif(len(T2)>=1 & T2[-1]==disc):
                    pos[disc]=0
                    T3.append(disc)
                    T2.remove(disc)
                    move[0] += 1
                    print(f'disc: {disc} position: {pos} move: {move} \n')
                    tour_hanoi(disc-1)
                else:
                    tour_hanoi(disc-1)
    else:
        print("error")
```
